```python
def cos_func(dff):
    cos_similarity = []
    df = dff.drop(['품명','시약_범주1','시약_범주2','시약_범주3'],axis=1)
    if df.shape[0] <= 1:
        return '단일 상품'
    
    for i in range(len(df)-1):
        for j in range(i+1, len(df)):
            simil_val = cos_sim(df.iloc[i],df.iloc[j])
            cos_similarity.append([simil_val,i,dff.품명.iloc[i],j,dff.품명.iloc[j]])
    return cos_similarity
```

```python
def cos_sim(A, B):
    return dot(A, B)/(norm(A)*norm(B))
```
